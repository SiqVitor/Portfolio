```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from re import split
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, confusion_matrix

sns.set_theme()
```

```python
# Read the data and display information
train_data = pd.read_csv('Data/[Kaggle] Titanic - Machine Learning from Disaster/train.csv')
validation_data = pd.read_csv('Data/[Kaggle] Titanic - Machine Learning from Disaster/test.csv')

# Display information about train_data
train_data.info()
```

#### Here we note that the data have 891 observations, 1 ID column, 1 target column (Suvived) and 10 features: 5 numerical and 5 categorical

```python
train_data.head(3)
```

```python
# Fill missing values in 'Cabin' column with 'Z0'
train_data['Cabin'].fillna('Z0', inplace=True)

# Split 'Cabin' column into 'Cabin_letter' and 'Cabin_number'
train_cabin_split = train_data['Cabin'].str.split(r'(\d+)', expand=True).iloc[:, :2]
train_cabin_split.columns = ['Cabin_letter', 'Cabin_number']

# Assign the new columns to 'train_data'
train_data[['Cabin_letter', 'Cabin_number']] = train_cabin_split

# Convert 'Cabin_number' column to numeric
train_data['Cabin_number'] = pd.to_numeric(train_data['Cabin_number'])

# Describing Data
train_data.describe()
```

```python
# Select numeric features from train_data
train_data_numeric = train_data.select_dtypes(include='number').drop(columns='PassengerId')

# Plot correlation heatmap
corr_matrix = train_data_numeric.corr()
sns.heatmap(corr_matrix, cmap='coolwarm')
plt.show()
```

```python
# Calculate class counts and proportions
class_counts = train_data['Survived'].value_counts()
class_proportions = class_counts / train_data.shape[0]

# Create class table DataFrame
class_table = pd.DataFrame({'Count': class_counts, 'Proportion': class_proportions})
class_table = class_table.sort_index()

print(class_table)
```

```python
# Plot histograms of numeric features in train_data
train_data.hist(figsize=(12, 10), bins=30)
plt.tight_layout()  # Ensures tight layout of subplots
plt.show()
```

```python
# Concatenate numeric features with one-hot encoded categorical features
categorical_features = ['Sex', 'Embarked', 'Cabin_letter']
train_data_cleaned = pd.concat([train_data_numeric, pd.get_dummies(train_data[categorical_features])], axis=1)

# Split data into features (X) and target (y)
X = train_data_cleaned.drop('Survived', axis=1)
y = train_data_cleaned['Survived']

# Fill missing values with the mean of each column
X.fillna(X.mean(), inplace=True)

# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)

# Initialize and train the random forest classifier
rfc = RandomForestClassifier(n_estimators=200, criterion='gini', random_state=42)
rfc.fit(X_train, y_train)

# Predict and calculate accuracy on the testing set
y_pred = rfc.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)

# Calculate and plot confusion matrix
cm = confusion_matrix(y_test, y_pred)
labels = ['Not Survived', 'Survived']
plt.figure(figsize=(6, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=labels, yticklabels=labels)
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
plt.show()
```

```python
## Fill missing values in 'Cabin' column with 'Z0'
validation_data['Cabin'].fillna('Z0', inplace=True)

# Split 'Cabin' column into 'Cabin_letter' and 'Cabin_number'
validation_cabin_split = validation_data['Cabin'].str.split(r'(\d+)', expand=True).iloc[:, :2]
validation_cabin_split.columns = ['Cabin_letter', 'Cabin_number']

# Assign the new columns to 'validation_data'
validation_data[['Cabin_letter', 'Cabin_number']] = validation_cabin_split

# Convert 'Cabin_number' column to numeric
validation_data['Cabin_number'] = pd.to_numeric(validation_data['Cabin_number'])

# Select numeric features from validation_data
validation_data_numeric = validation_data.select_dtypes(include='number').drop(columns='PassengerId')

# Concatenate numeric features with one-hot encoded categorical features and fill NA with means
validation_data_cleaned = pd.concat([validation_data_numeric, pd.get_dummies(validation_data[categorical_features])], axis=1)
validation_data_cleaned.fillna(X.mean(), inplace=True)
validation_data_cleaned = validation_data_cleaned.reindex(columns=X.columns, fill_value=0)

# Predict using the trained random forest classifier
val_predict = rfc.predict(validation_data_cleaned)

# Create submission DataFrame and save to a CSV file
submission = pd.DataFrame({'PassengerId': validation_data['PassengerId'], 'Survived': val_predict})
submission.to_csv('Data/[Kaggle] Titanic - Machine Learning from Disaster/submission.csv', index=False)
```
