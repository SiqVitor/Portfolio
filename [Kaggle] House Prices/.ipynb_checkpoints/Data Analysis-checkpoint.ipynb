```python
#importing libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#readind data
train, test = pd.read_csv('Data/train.csv'), pd.read_csv('Data/test.csv')
```

```python
#displaying relevant information about the data
string = '''The data has {rows} observations and {columns} variables.
The train and test proportion of entire data is {train_prop}% and {test_prop}%, respectively.'''

rows, columns = train.shape[0] + test.shape[0], train.shape[1]
train_prop, test_prop = np.round((train.shape[0] / rows * 100, test.shape[0] / rows * 100), 1)

print(string.format(rows = rows, columns = columns,
                    train_prop = train_prop, test_prop = test_prop))
```

```python
#identifying variables types and NaN presence
train.info()
```

```python
#cleaning variables with more than 20% of NaN values
na_limit = train.shape[0] * 0.2
cols_not_na = (train.isna().sum() <= na_limit).values
print('Columns removed because of to many NaN entries:\n{cols}'.format(cols = train.columns[~cols_not_na].values))
train, test = train.loc[:, cols_not_na], test.loc[:, cols_not_na[:-1]]
```
